## 4.2 Plotting activation functions 

### Relu, Sigmoid and their derivatives

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input range for plotting
x = np.linspace(-10, 10, 200)
```

```python
# Sigmoid and its derivative
y_sigmoid = 1 /  (1 + np.exp(-x))
y_sigmoid_derivative = y_sigmoid * (1 - y_sigmoid)
```

```python
# ReLU and its derivative
y_relu = np.maximum(0, x)
y_relu_derivative  = (x > 0).astype(float)
```

```python
# Plotting
plt.figure(figsize=(12, 8))

# ReLU vs Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, y_relu, label="ReLU", color = 'blue')
plt.plot(x, y_sigmoid, label="Sigmoid", color = 'green')
plt.title("ReLU vs Sigmoid")
plt.grid(True)
plt.legend()

# ReLU vs its derivative
plt.subplot(2, 2, 2)
plt.plot(x, y_relu, label="ReLU", color='blue')
plt.plot(x, y_relu_derivative, label = "ReLU'", color = 'red')
plt.title("ReLU and its Derivative")
plt.grid(True)
plt.legend()

# Sigmoid vs Sigmoid derivative
plt.subplot(2, 2, 3)
plt.plot(x, y_sigmoid, label="Sigmoid", color = 'green')
plt.plot(x, y_sigmoid_derivative, label="Sigmoid'", color = 'orange')
plt.title("Sigmoid vs Sigmoid derivative")
plt.grid(True)
plt.legend()


plt.tight_layout()
plt.show()
```
